```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
a={
    "names":["amazon","netflix","flipcart","aha"],
    "values":["today","tomorrwo","rajesh","maddi"]
}
df=pd.DataFrame(a)
obj=LabelEncoder()
df["names_n"]=obj.fit_transform(df["names"])
df["values_n"]=obj.fit_transform(df["values"])
df
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
a={
    "x":[10,20,30,40,50,60],
     "y":[100,200,300,400,500,600]
}
df=pd.DataFrame(a)
obj=MinMaxScaler(feature_range=(0,1))
df[["x","y"]]=obj.fit_transform(df[["x","y"]])
df
```

```python
import pandas as pd
import numpy as np
a={
    "names":[8323,7899,8999,8799],
     "rent":[2,3,4,200],
    "values":[2345,7899,6788,9997]
}
df=pd.DataFrame(a)
cond1=df["rent"]<6
print(df[cond1])
df["rent"]=np.where(cond1,0,1)
df
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
students = [
              [85, 'M', 'verygood'],
              [95, 'F', 'excellent'],
              [60, np.nan, 'good'],
              [np.nan, 'M', 'average'],
               [70, 'M', 'good'],
              [np.nan, np.nan, 'verygood'],
              [60, 'F', 'verygood'],
              [98, 'M', 'excellent']]
cols=["marks","gender","result"]
df=pd.DataFrame(students,columns=cols)
obj=SimpleImputer(strategy="mean")
x=df[["marks"]].values
df["marks"]=obj.fit_transform(x)
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
from numpy import asarray
a=([["amazon"],["netflix"],["aha"],["hello"]])
obj=OrdinalEncoder()
result=obj.fit_transform(a)
result
```

```python
from sklearn.preprocessing import OneHotEncoder
from numpy import asarray
x=asarray([["apple"],["horse"],["elephant"],["zink"]])
obj=OneHotEncoder(drop="first",sparse_output=False)
result=obj.fit_transform(x)
result
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer
students = [
              [85, 'M', 'verygood'],
              [95, 'F', 'excellent'],
              [60, np.nan, 'good'],
              [np.nan, 'M', 'average'],
               [70, 'M', 'good'],
              [np.nan, np.nan, 'verygood'],
              [60, 'F', 'verygood'],
              [98, 'M', 'excellent']]
cols=["marks","gender","result"]
df=pd.DataFrame(students,columns=cols)
obj=SimpleImputer(strategy="most_frequent")
x=df[["gender"]].values
df[["gender"]]=obj.fit_transform(x)
df
```
